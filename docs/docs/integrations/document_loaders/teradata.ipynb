## Overview
This example demonstrates how to use the Teradata document loader to load data from a Teradata Vantage database into LangChain documents. Each document corresponds to one row of query results:
- `page_content_columns`: These are written into the `page_content` of each document.
- `metadata_columns`: These are written into the `metadata` of each document. 
- By default, all columns are included in `page_content` and none in `metadata`.

```python
from langchain_community.document_loaders.teradata import TeradataLoader
```

```python
loader = TeradataLoader(
    query="SELECT comment_id, customer_id, comment_summary FROM DEMO_Retail.Web_Comment",
    db_url="HOST_URL",
    user="demo_user",
    password="password",
    page_content_columns=["comment_summary"],
    metadata_columns=["customer_id"]
)
```

```python
teradata_documents = list(loader.lazy_load())
print(teradata_documents)
# for document in teradata_documents:
#     print(document)

# # for metadata in teradata_documents:
# #     print(metadata)
```
